## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## My Lane as an ML Task

**Task Type:** Binary Classification

The Content Performance Prediction lane is a binary classification problem. The model predicts whether a content page is likely to experience a significant decline in search impressions over the next 30 days. Each page is classified into one of two classes: Declining or Not Declining. This helps content teams identify pages that require optimization before traffic drops.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

**Target:** Whether a content page's search impressions decline by more than 20% over the next 30 days.

The target label is created from historical search performance data. Pages with a decline greater than 20% are labeled as "Declining" (1), while the remaining pages are labeled as "Not Declining" (0).

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

The primary success metric is the F1-score because both false positives and false negatives are important. A higher F1-score indicates better performance in identifying pages that are likely to lose search traffic.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

## Unit of Analysis

Each row in the dataset represents one content page (identified by a unique content_hash_id). For every page, the dataset contains historical search performance features such as impressions, clicks, average position, and query-level signals. The model uses these features to predict whether that page is likely to lose search impressions in the future.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A simple if-statement cannot capture the complex relationships between impressions, clicks, rankings, and query behavior. Machine learning can learn patterns from many features at once and make better predictions than a single fixed rule. This allows more accurate identification of pages that may decline in search performance.
